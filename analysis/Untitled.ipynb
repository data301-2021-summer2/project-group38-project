```python
%matplotlib inline
sc = sns.scatterplot(data= df1, x="date", y="avg_temperature")
sc
```

```python
The plot above shows average temperatures from 2007-2020.
As seen in the plot above, average temperatures appear to remain relatively consistent throughout most of the recorded years, however there are distinct outliers. 
The outliers are representative of extreme temperatures. It is clear that in recent years there are more outliers within the hotter temperatures. 
```
